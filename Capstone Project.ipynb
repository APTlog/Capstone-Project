```python
def pasplot(feature):
    facet = sns.FacetGrid(train, hue = 'Absent', aspect =4)
    facet.map(sns.kdeplot, feature, shade = True)
    facet.set(xlim=(0, train[feature].max()))
    facet.add_legend()
```

```python
pasplot(' Work load Average/day  ')
```

```python
def Wload(train):
    if train[" Work load Average/day  "] <=215000:
        return "0 - 215000"
    elif (train[" Work load Average/day  "] <=268000):
        return "215001 - 268000"
    elif (train[" Work load Average/day  "] <=340000):
        return "268001 - 340000"
    elif (train[" Work load Average/day  "] >340000):
        return "Over 340000"
train["Wload"] = train.apply(lambda train:Wload(train), axis=1)
```
